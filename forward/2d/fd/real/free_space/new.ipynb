```python
dim_x,dim_y = 1024,1024
a_x,a_y = 3500,3500
```

```python
f_sol = h5py.File('solution.h5','r')
dset_sol = f_sol["sol_vec"]
dset_sol.shape
```

```python
i = 25
fr1 = np.zeros((dim_x,dim_y))
fr2 = np.zeros((dim_x,dim_y))
dset_sol.read_direct(fr1,np.s_[i,:,:,0], np.s_[:,:])
dset_sol.read_direct(fr2,np.s_[i,:,:,1], np.s_[:,:])
```

```python
plt.rcParams["figure.figsize"] = (12,5)
f,(ax1,ax2) = plt.subplots(1,2)

im1 = ax1.imshow(fr1,cmap="coolwarm")
f.colorbar(im1,ax=ax1)
im2 = ax2.imshow(fr2,cmap="coolwarm")
f.colorbar(im2,ax=ax2)
print(np.mean(fr1),np.mean(fr2))
```
